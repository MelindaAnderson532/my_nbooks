# Exercise 3-1: Create some plots

```python
import pandas as pd
```

## Get the data

```python
mortality_data = pd.read_pickle('mortality_prepped.pkl')
mortality_data.head()
```

```python
mortality_wide = pd.read_pickle('mortality_wide.pkl')
mortality_wide.head()
```

## Visualize the data

```python
mortality_data.query('AgeGroup == "01-04 Years"').plot.scatter(x='Year', y='DeathRate')
```

```python
mortality_wide.plot()
```

```python
mortality_wide.plot.line(title='Child Mortality: AgeGroup 15-19 Years', y='15-19 Years',legend=False)
```

```python
mortality_wide.plot.area(legend='reverse')
```

```python
mortality_wide.query('Year == (1900)').plot.bar(x='Year', ylabel="Mortality Rate",rot=0)
```

```python
mortality_wide.query('Year == (2000)').plot.bar(title='Child Mortality Rate for Year 2000',x='Year',xlabel='',rot=0)
```

```python
mortality_data.query('Year == (2000)').pivot(index='AgeGroup', columns='Year', values='DeathRate').plot.bar(title='Child Mortality Rate for 2000', legend=False, rot=0)
```

```python
mortality_data.plot.hist(y='DeathRate')
```

```python
mortality_data.plot.hist(y='DeathRate', bins=15) #The 15 bins are a little easier to read
```

```python
mortality_data.query('Year ==(1900)').plot.density(title='Child Mortality: 1900', y='DeathRate', grid=True)
```

```python
mortality_data.query('Year in (1900,1925,1950,1975,2000)').groupby('Year')['DeathRate'].sum().plot.pie()
```

```python
mortality_wide.query('Year in (1900,1925,1950,1975,2000)').plot.barh(title=['Child Mortality: 01-04', 'Child Mortality: 05-09',
                                'Child Mortality: 10-14', 'Child Mortality: 15-19'],
                         xlabel='Deaths per 100,000', sharey=True,
                         grid=True, rot=0, x='Year', legend=False,
                         subplots=True, layout=(2,2), figsize=(10,10))
```
